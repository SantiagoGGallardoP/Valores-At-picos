```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar archivo csv desde equipo
from google.colab import files
files.upload()
```

```python
#Carga desde un archivo .csv sin indice
data2020= pd.read_excel('G&C_LIMPIA.xlsx', sheet_name="2020") 
data2021= pd.read_excel('G&C_LIMPIA.xlsx', sheet_name="2021") 
data2022= pd.read_excel('G&C_LIMPIA.xlsx', sheet_name="2022") 
data2023= pd.read_excel('G&C_LIMPIA.xlsx', sheet_name="2023") 
```

#2020

```python
#Verificamos información del DataFrame
data2020.info()
```

```python
columnas_object = data2020.select_dtypes(exclude=['float', 'int'])
data_2020_str = pd.DataFrame(columnas_object)
columnas_numericas = data2020.select_dtypes(include=['float', 'int'])
data_2020_num = pd.DataFrame(columnas_numericas)
```

```python
#Corroboramos valores nulos
valores_nulos=data2020.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data_2020_num
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data_2020_num[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data_2020_num2= data_2020_num[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_2020_num2
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data_2020_num2.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean2020=data_2020_num2.copy()
data_clean2020=data_clean2020.fillna(round(data_2020_num2.mean(),1))
data_clean2020
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean2020.isnull().sum()
valores_nulos
```

```python
df_2020_concat = pd.concat([data_clean2020, data_2020_str], axis=1)
df_2020_concat.to_excel('G&C_2020_Desv.xlsx', index=False)
df_2020_concat
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data_2020_num

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data_2020_num[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data2020_iqr= data_2020_num[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2020_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data2020_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data2020_iqr=data2020_iqr.copy()
data2020_iqr=data2020_iqr.fillna(round(data2020_iqr.mean(),1))
data2020_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data2020_iqr.isnull().sum()
valores_nulos
```

```python
df_2020_concat_2 = pd.concat([data2020_iqr, data_2020_str], axis=1)
df_2020_concat_2.to_excel('G&C_2020_IQR.xlsx', index=False)
df_2020_concat_2
```
