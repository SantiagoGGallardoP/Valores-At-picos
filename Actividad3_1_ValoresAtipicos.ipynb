#ventas_precios_corrientes

```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize =(5, 3))
plt.hist(x=data["ventas_precios_corrientes"], color='blue', rwidth=0.50)
plt.title('Histograma de Precio corrientes')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(10, 6))
plt.boxplot(data["ventas_precios_corrientes"]) 
plt.title("Outliers de ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

##M√âTODO DESVIACI√ìN ESTANDAR 

```python
#M√©todo aplicando desviaci√≥n estandar. Encuentro los valores extremos
y=data["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= data[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["ventas_precios_corrientes"]) 
plt.title("Outliers de ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

##Rango Intercuart√≠lico

```python
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr_ventas_precios_corrientes= data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr_ventas_precios_corrientes
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr_ventas_precios_corrientes["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```
